```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Data Source- UCI Machine Learning Repository 
# https://archive.ics.uci.edu/ml/index.php
#creating dataframe
#column names are given explicitly
iris= pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',\
                  names=['Sepal Length','Sepal Width','Petal Length','Petal Width','Class'])
```

```python
iris.head()
```

```python
#visualizing data
iris.plot(kind='scatter',x='Petal Length',y='Petal Width')
```

Using above scatter plot we are unable to differentiate between classes of iris flower.

```python
sns.scatterplot(data= iris, x='Petal Length',y='Petal Width', hue='Class')
plt.show()
```

From above scatter plot, we can infer that there are separate clusters of different classes of iris flower.

```python
#There are total 5 attributes in iris table. Four attributes will remain if we use Class attribute 
#as hue.
#Therefore, total graph possible are 4*4= 16.
sns.pairplot(data=iris, hue='Class')
```
